```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
#loading raw noise data 
sound1 = pd.read_csv('as10724_rms.csv')
sound1.describe()
```

```python
#Visiualizing the raw data
plt.figure(figsize=(18,8))
plt.plot(sound1['Time2'], sound1['RMS'], 'k', alpha=.8)
plt.title('Noise in RMS', size=24)
plt.xlabel("Time in seconds", size=14)
plt.ylabel("RMS", size=14)
```

```python
#loading aggregated data in decibels 
sound2 = pd.read_csv('as10724_decibel.csv')
sound2.describe()
```

```python
#setting datetime index
sound2.index = pd.to_datetime(sound2.Time)
```

```python
#Visiualizing the raw data
plt.figure(figsize=(18,8))
plt.plot(sound2.index, sound2['decibels'], 'k', alpha=.8)
plt.plot([sound2.index[0], sound2.index[-1]], [np.mean(sound2.decibels), np.mean(sound2.decibels)], 'r', label = 'Mean')
plt.title('Noise in decibels', size=24)
plt.xlabel("Time", size=14)
plt.ylabel("decibels", size=14)
plt.legend()
```
